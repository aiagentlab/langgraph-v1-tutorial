### ğŸ“‹ í—¬í¼ í•¨ìˆ˜ì˜ ì—­í• 

ë©€í‹°ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œì—ì„œëŠ” ì—¬ëŸ¬ ì—ì´ì „íŠ¸ê°€ ë™ì‹œì— ì‘ì—…í•˜ë¯€ë¡œ, ê° ì—ì´ì „íŠ¸ì˜ ì¶œë ¥ì„ êµ¬ë¶„í•´ì„œ ë³´ê¸° ì–´ë ¤ìš¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ë¥¼ í•´ê²°í•˜ê¸° ìœ„í•´ ë‹¤ìŒ í•¨ìˆ˜ë“¤ì„ ì‚¬ìš©í•©ë‹ˆë‹¤:

- **`pretty_print_message()`**: ê°œë³„ ë©”ì‹œì§€ë¥¼ ë³´ê¸° ì¢‹ê²Œ í¬ë§·íŒ…
- **`pretty_print_messages()`**: ì—¬ëŸ¬ ì—ì´ì „íŠ¸ì˜ ë©”ì‹œì§€ ì—…ë°ì´íŠ¸ë¥¼ êµ¬ë¶„í•˜ì—¬ í‘œì‹œ
- ê° ë…¸ë“œ(ì—ì´ì „íŠ¸)ë³„ë¡œ ìƒ‰ìƒê³¼ êµ¬ë¶„ì„ ì„ ì¶”ê°€í•˜ì—¬ ê°€ë…ì„± í–¥ìƒ

````python
def pretty_print_message(message, indent=False):
    """Format and print a single message"""
    pretty_message = message.pretty_repr(html=True)
    if not indent:
        print(pretty_message)
        return

    # ë“¤ì—¬ì“°ê¸° ì ìš©
    indented = "\n".join("\t" + c for c in pretty_message.split("\n"))
    print(indented)


def pretty_print_messages(update, last_message=False):
    """Format and print message updates from graph execution"""
    is_subgraph = False

    # íŠœí”Œì¸ ê²½ìš° ì„œë¸Œê·¸ë˜í”„ ì—…ë°ì´íŠ¸
    if isinstance(update, tuple):
        ns, update = update
        # ë¶€ëª¨ ê·¸ë˜í”„ ì—…ë°ì´íŠ¸ëŠ” ê±´ë„ˆëœ€
        if len(ns) == 0:
            return

        graph_id = ns[-1].split(":")[0]
        print(f"ğŸ“ ì„œë¸Œê·¸ë˜í”„ {graph_id}ë¡œë¶€í„° ì—…ë°ì´íŠ¸:")
        print("\n")
        is_subgraph = True

    # ê° ë…¸ë“œì˜ ì—…ë°ì´íŠ¸ ì²˜ë¦¬
    for node_name, node_update in update.items():
        update_label = f"ğŸ“Œ ë…¸ë“œ {node_name}ë¡œë¶€í„° ì—…ë°ì´íŠ¸:"
        if is_subgraph:
            update_label = "\t" + update_label

        print(update_label)
        print("\n")

        # ë©”ì‹œì§€ ë³€í™˜ ë° ì¶œë ¥
        messages = convert_to_messages(node_update["messages"])
        if last_message:
            messages = messages[-1:]  # ë§ˆì§€ë§‰ ë©”ì‹œì§€ë§Œ

        for m in messages:
            pretty_print_message(m, indent=is_subgraph)
        print("\n")


print("âœ… í—¬í¼ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ!")
````

## 1.1 Research Agent (ì—°êµ¬ ì—ì´ì „íŠ¸)

Tavily APIë¥¼ ì‚¬ìš©í•˜ì—¬ ì›¹ ê²€ìƒ‰ ê¸°ëŠ¥ì„ ê°–ì¶˜ ì—°êµ¬ ì—ì´ì „íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

````python
from langchain_tavily import TavilySearch
from langgraph.prebuilt import create_react_agent

# Tavily ì›¹ ê²€ìƒ‰ ë„êµ¬ ì´ˆê¸°í™” (ìµœëŒ€ 3ê°œ ê²°ê³¼ ë°˜í™˜)
web_search = TavilySearch(max_results=3)

# ì›¹ ê²€ìƒ‰ ë„êµ¬ í…ŒìŠ¤íŠ¸ ì‹¤í–‰
test_results = web_search.invoke("2024ë…„ ëŒ€í•œë¯¼êµ­ GDP")
print("ğŸ” ì›¹ ê²€ìƒ‰ ë„êµ¬ í…ŒìŠ¤íŠ¸:")
print(f"ê²€ìƒ‰ ê²°ê³¼ ìˆ˜: {len(test_results['results'])}")
print(f"ì²« ë²ˆì§¸ ê²°ê³¼ ë¯¸ë¦¬ë³´ê¸°: {test_results['results'][0]['content'][:200]}...\n")

# Research Agent ìƒì„± (ì›¹ ê²€ìƒ‰ ì „ë¬¸ ì—ì´ì „íŠ¸)
research_agent = create_react_agent(
    model="gpt-4.1",  # ì‚¬ìš©í•  LLM ëª¨ë¸
    tools=[web_search],  # ì›¹ ê²€ìƒ‰ ë„êµ¬ ì œê³µ
    prompt=(
        "You are a research specialist agent.\n\n"
        "CORE RESPONSIBILITIES:\n"
        "- Conduct thorough web searches to find accurate and up-to-date information\n"
        "- Provide detailed, factual responses based on search results\n"
        "- Always cite your sources when presenting information\n\n"
        "STRICT GUIDELINES:\n"
        "- Focus ONLY on research and information gathering tasks\n"
        "- DO NOT perform mathematical calculations or analysis\n"
        "- When task is complete, report findings directly to the supervisor\n"
        "- Present ONLY the research results without additional commentary\n\n"
        "OUTPUT FORMAT:\n"
        "- Provide clear, structured findings\n"
        "- Include relevant data points and sources\n"
        "- Use bullet points for clarity when appropriate"
    ),
    name="research_agent",  # ì—ì´ì „íŠ¸ ì´ë¦„
)

print("âœ… Research Agent ìƒì„± ì™„ë£Œ!")
````

````python
# Research Agent í…ŒìŠ¤íŠ¸
print("ğŸ§ª Research Agent í…ŒìŠ¤íŠ¸\n")
print("ì§ˆë¬¸: 'NYCì˜ í˜„ì¬ ì‹œì¥ì€ ëˆ„êµ¬ì¸ê°€ìš”?'\n")

for chunk in research_agent.stream(
    {"messages": [{"role": "user", "content": "NYCì˜ í˜„ì¬ ì‹œì¥ì€ ëˆ„êµ¬ì¸ê°€ìš”?"}]}
):
    pretty_print_messages(chunk)
````

## 1.2 Math Agent (ìˆ˜í•™ ì—ì´ì „íŠ¸)

ê¸°ë³¸ì ì¸ ìˆ˜í•™ ì—°ì‚°ì„ ìˆ˜í–‰í•  ìˆ˜ ìˆëŠ” ì—ì´ì „íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

````python
# ìˆ˜í•™ ë„êµ¬ ì •ì˜ (ê¸°ë³¸ ê³„ì‚° í•¨ìˆ˜ë“¤)
def add(a: float, b: float) -> float:
    """Add two numbers together."""
    return a + b


def multiply(a: float, b: float) -> float:
    """Multiply two numbers."""
    return a * b


def divide(a: float, b: float) -> float:
    """Divide two numbers. Returns error message if dividing by zero."""
    if b == 0:
        return float("inf")  # 0ìœ¼ë¡œ ë‚˜ëˆ„ê¸° ë°©ì§€, ë¬´í•œëŒ€ ë°˜í™˜
    return a / b


def power(base: float, exponent: float) -> float:
    """Calculate base raised to the power of exponent."""
    return base**exponent


def percentage(value: float, total: float) -> float:
    """Calculate what percentage 'value' is of 'total'."""
    if total == 0:
        return 0.0  # ì „ì²´ê°’ì´ 0ì¸ ê²½ìš° 0% ë°˜í™˜
    return (value / total) * 100


# Math Agent ìƒì„± (ìˆ˜í•™ ê³„ì‚° ì „ë¬¸ ì—ì´ì „íŠ¸)
math_agent = create_react_agent(
    model="gpt-4.1",  # ì‚¬ìš©í•  LLM ëª¨ë¸
    tools=[add, multiply, divide, power, percentage],  # ìˆ˜í•™ ë„êµ¬ë“¤ ì œê³µ
    prompt=(
        "You are a mathematics specialist agent.\n\n"
        "CORE RESPONSIBILITIES:\n"
        "- Perform accurate mathematical calculations\n"
        "- Handle arithmetic operations, percentages, and basic algebra\n"
        "- Show calculation steps when solving complex problems\n\n"
        "STRICT GUIDELINES:\n"
        "- Focus ONLY on mathematical computations and analysis\n"
        "- DO NOT perform web searches or gather external information\n"
        "- When calculations are complete, report results to the supervisor\n"
        "- Present ONLY the calculation results and methodology\n\n"
        "OUTPUT FORMAT:\n"
        "- Show clear calculation steps\n"
        "- Round results to 2 decimal places when appropriate\n"
        "- Include units if relevant"
    ),
    name="math_agent",  # ì—ì´ì „íŠ¸ ì´ë¦„
)

print("âœ… Math Agent ìƒì„± ì™„ë£Œ!")
````

````python
# Math Agent í…ŒìŠ¤íŠ¸
print("ğŸ§ª Math Agent í…ŒìŠ¤íŠ¸\n")
print("ì§ˆë¬¸: '(3 + 5) x 7ì€ ì–¼ë§ˆì¸ê°€ìš”?'\n")

for chunk in math_agent.stream(
    {"messages": [{"role": "user", "content": "(3 + 5) x 7ì€ ì–¼ë§ˆì¸ê°€ìš”?"}]}
):
    pretty_print_messages(chunk)
````

# Part 2: langgraph-supervisorë¡œ Supervisor ìƒì„± ğŸ¯

ì´ì œ ì‚¬ì „ êµ¬ì¶•ëœ `langgraph-supervisor` ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ supervisorë¥¼ ë¹ ë¥´ê²Œ êµ¬í˜„í•´ë´…ì‹œë‹¤.

````python
from langgraph_supervisor import create_supervisor
from langchain.chat_models import init_chat_model

# Supervisor ìƒì„± - ì‚¬ì „ êµ¬ì¶•ëœ íŒ¨í„´ ì‚¬ìš©
supervisor = create_supervisor(
    model=init_chat_model("gpt-4.1"),  # Supervisorê°€ ì‚¬ìš©í•  LLM ëª¨ë¸
    agents=[research_agent, math_agent],  # ê´€ë¦¬í•  ì—ì´ì „íŠ¸ë“¤ (ì—°êµ¬, ìˆ˜í•™)
    prompt=(
        "You are a supervisor managing two specialized agents:\n\n"
        "1. RESEARCH AGENT:\n"
        "   - Specializes in web searches and information gathering\n"
        "   - Use for: finding facts, current events, statistics, definitions\n\n"
        "2. MATH AGENT:\n"
        "   - Specializes in mathematical calculations and analysis\n"
        "   - Use for: arithmetic, percentages, conversions, comparisons\n\n"
        "ORCHESTRATION GUIDELINES:\n"
        "- Analyze the user's request carefully\n"
        "- Delegate tasks to the appropriate specialist agent\n"
        "- Process tasks sequentially (one agent at a time)\n"
        "- Coordinate multiple agents when complex tasks require both research and calculation\n"
        "- Synthesize results from agents into a coherent final response\n\n"
        "IMPORTANT RULES:\n"
        "- DO NOT perform any work yourself - always delegate to agents\n"
        "- DO NOT call agents in parallel - use sequential processing\n"
        "- Always provide clear, complete answers based on agent results"
    ),
    add_handoff_back_messages=True,  # ì—ì´ì „íŠ¸ ê°„ í•¸ë“œì˜¤í”„ ë©”ì‹œì§€ ì¶”ê°€
    output_mode="full_history",  # ì „ì²´ íˆìŠ¤í† ë¦¬ ì¶œë ¥ ëª¨ë“œ
).compile()  # ê·¸ë˜í”„ ì»´íŒŒì¼

print("âœ… Supervisor ìƒì„± ì™„ë£Œ!")
````

## 2.1 Supervisor í…ŒìŠ¤íŠ¸ - ë³µí•© ì‘ì—…

Researchì™€ Math ì—ì´ì „íŠ¸ê°€ ëª¨ë‘ í•„ìš”í•œ ë³µì¡í•œ ì§ˆë¬¸ìœ¼ë¡œ í…ŒìŠ¤íŠ¸í•´ë´…ì‹œë‹¤.

````python
# ë³µí•© ì‘ì—… í…ŒìŠ¤íŠ¸: GDP ì •ë³´ ê²€ìƒ‰ + ë¹„ìœ¨ ê³„ì‚°
print("ğŸ”¬ Supervisor ë³µí•© ì‘ì—… í…ŒìŠ¤íŠ¸\n")
print(
    "ì§ˆë¬¸: '2024ë…„ ë¯¸êµ­ê³¼ ë‰´ìš•ì£¼ì˜ GDPë¥¼ ì°¾ê³ , ë‰´ìš•ì£¼ê°€ ë¯¸êµ­ ì „ì²´ GDPì˜ ëª‡ %ë¥¼ ì°¨ì§€í•˜ëŠ”ì§€ ê³„ì‚°í•´ì£¼ì„¸ìš”.'\n"
)
print("=" * 80 + "\n")

# Supervisor ì‹¤í–‰
for chunk in supervisor.stream(
    {
        "messages": [
            {
                "role": "user",
                "content": "2024ë…„ ë¯¸êµ­ê³¼ ë‰´ìš•ì£¼ì˜ GDPë¥¼ ì°¾ê³ , ë‰´ìš•ì£¼ê°€ ë¯¸êµ­ ì „ì²´ GDPì˜ ëª‡ %ë¥¼ ì°¨ì§€í•˜ëŠ”ì§€ ê³„ì‚°í•´ì£¼ì„¸ìš”.",
            }
        ]
    },
):
    pretty_print_messages(chunk, last_message=True)

# ìµœì¢… ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬ ì €ì¥
final_message_history = chunk.get("supervisor", {}).get("messages", [])
````

````python
# ì „ì²´ ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬ í™•ì¸
print("ğŸ“œ ì „ì²´ ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬:\n")
print("=" * 80 + "\n")

for i, message in enumerate(final_message_history, 1):
    print(f"[ë©”ì‹œì§€ {i}]")
    message.pretty_print()
    print("-" * 40)
````

## 3.1 ì—ì´ì „íŠ¸ ê°„ í†µì‹  ì„¤ì • (Handoff)

**í•¸ë“œì˜¤í”„(Handoff)** ëŠ” í•œ ì—ì´ì „íŠ¸ê°€ ë‹¤ë¥¸ ì—ì´ì „íŠ¸ì—ê²Œ ì œì–´ê¶Œì„ ë„˜ê¸°ëŠ” ë©”ì»¤ë‹ˆì¦˜ì…ë‹ˆë‹¤.

### ğŸ¤ í•¸ë“œì˜¤í”„ë€?

ì‹¤ì œ ì—…ë¬´ì—ì„œ **"ì´ ì—…ë¬´ëŠ” AíŒ€ì—ì„œ ì²˜ë¦¬í•˜ê³ , ì™„ë£Œë˜ë©´ BíŒ€ìœ¼ë¡œ ë„˜ê²¨ì£¼ì„¸ìš”"** ë¼ê³  í•˜ëŠ” ê²ƒê³¼ ê°™ì€ ê°œë…ì…ë‹ˆë‹¤.

**ì˜ˆì‹œ:**
1. ê³ ê°ì´ "ì˜¬í•´ ì• í”Œ ì£¼ê°€ ë™í–¥ì„ ë¶„ì„í•´ì£¼ì„¸ìš”"ë¼ê³  ìš”ì²­
2. **Supervisor**: "ë¨¼ì € ì• í”Œ ì£¼ê°€ ë°ì´í„°ë¥¼ ì¡°ì‚¬í•´ì•¼ê² ë„¤" â†’ **Research Agent** ë¡œ í•¸ë“œì˜¤í”„
3. **Research Agent**: ì›¹ì—ì„œ ì• í”Œ ì£¼ê°€ ì •ë³´ ìˆ˜ì§‘ ì™„ë£Œ â†’ **Supervisor** ë¡œ ë³µê·€
4. **Supervisor**: "ì´ì œ ìˆ˜ì§‘ëœ ë°ì´í„°ë¡œ ê³„ì‚°í•´ë³´ì" â†’ **Math Agent** ë¡œ í•¸ë“œì˜¤í”„

### ğŸ”§ í•¸ë“œì˜¤í”„ êµ¬í˜„ ë°©ì‹

LangGraphì—ì„œëŠ” **Command** ê°ì²´ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¤ìŒ ì—ì´ì „íŠ¸ë¡œ ì œì–´ê¶Œì„ ì „ë‹¬í•©ë‹ˆë‹¤.

````python
from langchain_core.tools import tool, InjectedToolCallId
from langgraph.prebuilt import InjectedState
from langgraph.graph import StateGraph, START, MessagesState
from langgraph.types import Command


def create_handoff_tool(*, agent_name: str, description: str | None = None):
    """Create a tool for handing off control to another agent"""
    # ë„êµ¬ ì´ë¦„ ìƒì„±
    name = f"transfer_to_{agent_name}"
    description = (
        description or f"Transfer task to {agent_name} for specialized processing."
    )

    @tool(name, description=description)
    def handoff_tool(
        state: Annotated[MessagesState, InjectedState],  # í˜„ì¬ ìƒíƒœ ì£¼ì…
        tool_call_id: Annotated[str, InjectedToolCallId],  # ë„êµ¬ í˜¸ì¶œ ID ì£¼ì…
    ) -> Command:
        """Execute handoff to target agent"""
        # í•¸ë“œì˜¤í”„ ì„±ê³µ ë©”ì‹œì§€ ìƒì„±
        tool_message = {
            "role": "tool",
            "content": f"âœ… Successfully transferred to {agent_name}",
            "name": name,
            "tool_call_id": tool_call_id,
        }

        # Commandë¥¼ í†µí•´ ë‹¤ìŒ ì—ì´ì „íŠ¸ë¡œ ë¼ìš°íŒ…
        return Command(
            goto=agent_name,  # ëŒ€ìƒ ì—ì´ì „íŠ¸ ì§€ì •
            update={  # ìƒíƒœ ì—…ë°ì´íŠ¸
                **state,
                "messages": state["messages"] + [tool_message],
            },
            graph=Command.PARENT,  # ë¶€ëª¨ ê·¸ë˜í”„ì—ì„œ ë„¤ë¹„ê²Œì´ì…˜
        )

    return handoff_tool


# ê° ì—ì´ì „íŠ¸ë¡œì˜ í•¸ë“œì˜¤í”„ ë„êµ¬ ìƒì„±
assign_to_research_agent = create_handoff_tool(
    agent_name="research_agent",
    description="Assign information gathering and web search tasks to the research specialist.",
)

assign_to_math_agent = create_handoff_tool(
    agent_name="math_agent",
    description="Assign mathematical calculations and numerical analysis to the math specialist.",
)

print("âœ… í•¸ë“œì˜¤í”„ ë„êµ¬ ìƒì„± ì™„ë£Œ!")
print(f"  - {assign_to_research_agent.name}: Research ì—ì´ì „íŠ¸ë¡œ ì‘ì—… ìœ„ì„")
print(f"  - {assign_to_math_agent.name}: Math ì—ì´ì „íŠ¸ë¡œ ì‘ì—… ìœ„ì„")
````

## 3.2 Supervisor ì—ì´ì „íŠ¸ ìƒì„±

````python
# ì»¤ìŠ¤í…€ Supervisor ì—ì´ì „íŠ¸ ìƒì„± (í•¸ë“œì˜¤í”„ ë„êµ¬ í™œìš©)
supervisor_agent = create_react_agent(
    model="gpt-4.1",  # ì‚¬ìš©í•  LLM ëª¨ë¸
    tools=[assign_to_research_agent, assign_to_math_agent],  # í•¸ë“œì˜¤í”„ ë„êµ¬ë“¤
    prompt=(
        "You are a supervisor orchestrating specialized agents.\n\n"
        "AVAILABLE AGENTS:\n"
        "1. Research Agent: Web searches, fact-finding, information gathering\n"
        "2. Math Agent: Calculations, percentages, numerical analysis\n\n"
        "ORCHESTRATION STRATEGY:\n"
        "- Analyze requests to identify required capabilities\n"
        "- Delegate to appropriate agents using transfer tools\n"
        "- Coordinate multi-step tasks by sequencing agent calls\n"
        "- Synthesize results into comprehensive responses\n\n"
        "EXECUTION RULES:\n"
        "- Always delegate work to specialist agents\n"
        "- Process tasks sequentially (one agent at a time)\n"
        "- Wait for agent results before proceeding\n"
        "- Provide clear final answers based on collected information"
    ),
    name="supervisor",  # ì—ì´ì „íŠ¸ ì´ë¦„
)

print("âœ… Supervisor ì—ì´ì „íŠ¸ ìƒì„± ì™„ë£Œ!")
````

## 3.3 ë©€í‹°ì—ì´ì „íŠ¸ ê·¸ë˜í”„ ìƒì„±

````python
from langgraph.graph import END

# ë©€í‹°ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œ ê·¸ë˜í”„ ì •ì˜
custom_supervisor = (
    StateGraph(MessagesState)
    # Supervisor ë…¸ë“œ ì¶”ê°€ (ëª©ì ì§€ íŒíŠ¸ í¬í•¨ - ì‹œê°í™”ìš©)
    .add_node(supervisor_agent, destinations=("research_agent", "math_agent", END))
    # ì›Œì»¤ ì—ì´ì „íŠ¸ ë…¸ë“œ ì¶”ê°€
    .add_node(research_agent)
    .add_node(math_agent)
    # ì‹œì‘ì ì—ì„œ Supervisorë¡œ
    .add_edge(START, "supervisor")
    # ì›Œì»¤ ì—ì´ì „íŠ¸ëŠ” í•­ìƒ Supervisorë¡œ ë³µê·€
    .add_edge("research_agent", "supervisor")
    .add_edge("math_agent", "supervisor")
    # ê·¸ë˜í”„ ì»´íŒŒì¼
    .compile()
)

print("âœ… ì»¤ìŠ¤í…€ ë©€í‹°ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œ ìƒì„± ì™„ë£Œ!")
````

# Part 4: ê³ ê¸‰ ì‘ì—… ìœ„ì„ (Task Delegation) ğŸ“‹

ì§€ê¸ˆê¹Œì§€ ì—ì´ì „íŠ¸ë“¤ì€ ì „ì²´ ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬ë¥¼ í•´ì„í•˜ì—¬ ì‘ì—…ì„ ìˆ˜í–‰í–ˆìŠµë‹ˆë‹¤.

### ğŸ¯ ê¸°ì¡´ ë°©ì‹ vs ê³ ê¸‰ ë°©ì‹ ë¹„êµ

#### ğŸŸ¢ ê³ ê¸‰ ë°©ì‹ (ëª…ì‹œì  ìœ„ì„)

### ğŸ’¡ ê³ ê¸‰ ë°©ì‹ì˜ ì¥ì 

- **ì •í™•ì„± í–¥ìƒ**: ì—ì´ì „íŠ¸ê°€ ì •í™•íˆ ë¬´ì—‡ì„ í•´ì•¼ í•˜ëŠ”ì§€ ëª…í™•íˆ ì•Œ ìˆ˜ ìˆìŒ
- **íš¨ìœ¨ì„± ì¦ëŒ€**: ë¶ˆí•„ìš”í•œ ì¶”ë¡  ê³¼ì • ì—†ì´ ë°”ë¡œ ì‘ì—… ìˆ˜í–‰
- **ì¼ê´€ì„± ë³´ì¥**: ë™ì¼í•œ ìœ í˜•ì˜ ì‘ì—…ì— ëŒ€í•´ í‘œì¤€í™”ëœ ì§€ì‹œ ê°€ëŠ¥

````python
from langgraph.types import Send


def create_task_description_handoff_tool(
    *, agent_name: str, description: str | None = None
):
    """Create a handoff tool that includes explicit task description"""
    name = f"transfer_to_{agent_name}"
    description = (
        description or f"Transfer task with detailed instructions to {agent_name}."
    )

    @tool(name, description=description)
    def handoff_tool(
        # Supervisorê°€ ì‘ì„±í•œ ì‘ì—… ì„¤ëª…
        task_description: Annotated[
            str,
            "Detailed description of what the agent should do, including all context and requirements.",
        ],
        # ìƒíƒœ ì£¼ì… (LLMì€ ë¬´ì‹œ)
        state: Annotated[MessagesState, InjectedState],
    ) -> Command:
        """Execute handoff with explicit task description"""
        # ì‘ì—… ì„¤ëª…ì„ ì‚¬ìš©ì ë©”ì‹œì§€ë¡œ ë³€í™˜
        task_description_message = {"role": "user", "content": task_description}
        # ì—ì´ì „íŠ¸ì—ê²Œ ì „ë‹¬í•  ìƒˆë¡œìš´ ìƒíƒœ ìƒì„±
        agent_input = {**state, "messages": [task_description_message]}

        return Command(
            # Sendë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¹ì • ìƒíƒœë¥¼ ì—ì´ì „íŠ¸ì—ê²Œ ì „ë‹¬
            goto=[Send(agent_name, agent_input)],
            graph=Command.PARENT,
        )

    return handoff_tool


# ì‘ì—… ì„¤ëª…ì´ í¬í•¨ëœ í•¸ë“œì˜¤í”„ ë„êµ¬ ìƒì„±
assign_research_with_description = create_task_description_handoff_tool(
    agent_name="research_agent",
    description="Assign research task with specific instructions.",
)

assign_math_with_description = create_task_description_handoff_tool(
    agent_name="math_agent",
    description="Assign calculation task with specific instructions.",
)

print("âœ… ì‘ì—… ì„¤ëª… í•¸ë“œì˜¤í”„ ë„êµ¬ ìƒì„± ì™„ë£Œ!")
````

````python
# ì‘ì—… ì„¤ëª…ì„ ì‚¬ìš©í•˜ëŠ” ê³ ê¸‰ Supervisor ìƒì„±
supervisor_with_delegation = create_react_agent(
    model="gpt-4.1",  # ì‚¬ìš©í•  LLM ëª¨ë¸
    tools=[
        assign_research_with_description,
        assign_math_with_description,
    ],  # ì‘ì—… ì„¤ëª… í•¸ë“œì˜¤í”„ ë„êµ¬ë“¤
    prompt=(
        "You are an advanced supervisor that delegates tasks with explicit instructions.\n\n"
        "DELEGATION STRATEGY:\n"
        "When assigning tasks to agents, provide:\n"
        "1. Clear, specific instructions about what to find or calculate\n"
        "2. All necessary context and parameters\n"
        "3. Expected output format or requirements\n"
        "4. Any constraints or special considerations\n\n"
        "AGENT CAPABILITIES:\n"
        "- Research Agent: Web searches, data gathering, fact-checking\n"
        "- Math Agent: All mathematical operations and analysis\n\n"
        "WORKFLOW RULES:\n"
        "- Create detailed task descriptions for each delegation\n"
        "- Process complex requests step-by-step\n"
        "- Ensure each agent has all information needed to succeed\n"
        "- Synthesize results into comprehensive final answers"
    ),
    name="supervisor",  # ì—ì´ì „íŠ¸ ì´ë¦„
)

# ê³ ê¸‰ Supervisor ì‹œìŠ¤í…œ ê·¸ë˜í”„ ìƒì„±
advanced_supervisor = (
    StateGraph(MessagesState)
    .add_node(
        supervisor_with_delegation,
        destinations=("research_agent", "math_agent"),  # ê°€ëŠ¥í•œ ëª©ì ì§€ ë…¸ë“œë“¤
    )
    .add_node(research_agent)  # ì—°êµ¬ ì—ì´ì „íŠ¸ ë…¸ë“œ ì¶”ê°€
    .add_node(math_agent)  # ìˆ˜í•™ ì—ì´ì „íŠ¸ ë…¸ë“œ ì¶”ê°€
    .add_edge(START, "supervisor")  # ì‹œì‘ì ì—ì„œ supervisorë¡œ ì—°ê²°
    .add_edge("research_agent", "supervisor")  # ì—°êµ¬ ì—ì´ì „íŠ¸ì—ì„œ supervisorë¡œ ë³µê·€
    .add_edge("math_agent", "supervisor")  # ìˆ˜í•™ ì—ì´ì „íŠ¸ì—ì„œ supervisorë¡œ ë³µê·€
    .compile()  # ê·¸ë˜í”„ ì»´íŒŒì¼
)

print("âœ… ê³ ê¸‰ Supervisor ì‹œìŠ¤í…œ ìƒì„± ì™„ë£Œ!")
````
